```python
from sklearn.cluster import KMeans
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
X= genfromtxt('/home/qmul/2D.csv', delimiter=',')
type(X)
```

```python
plt.scatter(X[:,1], X[:,2], c=X[:,0], s=5)
plt.ylabel('Number of posts by the user')
plt.xlabel('User reputation score')

yellow_patch = mpatches.Patch(color='khaki', label='The third cluster')
green_patch = mpatches.Patch(color='cadetblue', label='The second cluster')
purple_patch = mpatches.Patch(color='indigo', label='The fisrt cluster')
plt.legend(handles=[yellow_patch, green_patch, purple_patch])

plt.show()
```
